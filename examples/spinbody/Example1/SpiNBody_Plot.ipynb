```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

Time, dt, X, Y, Z, Vx, Vy, Vz = np.loadtxt('SpiNBody.Earth.forward', unpack=True)
sTime, sdt, sX, sY, sZ, sVx, sVy, sVz, AngMom, Energy = np.loadtxt('SpiNBody.Star.forward', unpack=True)
vTime, vdt, vX, vY, vZ, vVx, vVy, vVz = np.loadtxt('SpiNBody.Venus.forward', unpack=True)
mTime, mdt, mX, mY, mZ, mVx, mVy, mVz = np.loadtxt('SpiNBody.Mars.forward', unpack=True)
merTime, merdt, merX, merY, merZ, merVx, merVy, merVz = np.loadtxt('SpiNBody.Mercury.forward', unpack=True)
```

```python
plt.figure(figsize=[12,12])
plt.plot(merX,merY,'.k')
plt.plot(vX,vY,'.y')
plt.plot(X,Y,'.b')
plt.plot(mX,mY,'.r')
#plt.figure(figsize=[12,8])
plt.plot(sX,sY,'.m')
Timeyrs = max(Time/(3600*24*365.25))
plt.title("The Inner Solar System: %iyr simulation" %Timeyrs, fontsize=20)
plt.xlabel("X Position (m)", fontsize=16)
plt.ylabel("Y Position (m)", fontsize=16)
plt.legend(["Mercury","Venus","Earth","Mars"],prop={'size':16}, markerscale=3)
plt.savefig("Inner_Solar_System")
```

```python
def FindTimeStep(X,Y,Z,Vx,Vy,Vz):
    Distance = np.sqrt(X**2+Y**2+Z**2)
    Velocity = np.sqrt(Vx**2+Vy**2+Vz**2)
    return(min(Distance/Velocity))

TimeEarth = FindTimeStep(X,Y,Z,Vx,Vy,Vz)
TimeSun   = FindTimeStep(sX,sY,sZ,sVx,sVy,sVz)
TimeMars  = FindTimeStep(mX,mY,mZ,mVx,mVy,mVz)
TimeMercury = FindTimeStep(merX,merY,merZ,merVx,merVy,merVz)
TimeVenus = FindTimeStep(vX,vY,vZ,vVx,vVy,vVz)
print(TimeEarth*0.05)
print(TimeSun*0.05)
print(TimeMars*0.05)
print(TimeMercury*0.05)
print(TimeVenus*0.05)
```
